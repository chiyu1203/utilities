```python
#old_file='Y:\GN25051\251101\coherence\session1\camera2_2025-11-01T21_03_08.mp4'
for root, dirs, files in os.walk(thisDataset):
    for name in files:
        if name.endswith(file_type):
            if overwrite_existing_analysis==False and any(other_file.endswith(fictrac_output) for other_file in os.listdir(root)):
                print("fictrac .dat file exists and chose not to overwrite this file so move on to the next session")
                continue
            else:
                this_file=os.path.join(root,name)
                found_result=find_file(root,"camera*.dat")
                if found_result is None:
                    print("did not find any old dat file")
                    pass
                elif isinstance(found_result, list):
                    for old_file in found_result:
                        try:
                            Path.unlink(old_file)
                            print(f"remove old dat file {old_file}")
                        except OSError as e:
                            print("Error: %s - %s." % (e.filename, e.strerror))
                elif found_result.is_file() == True:
                    try:
                        Path.unlink(found_result)
                        print(f"remove old dat file {found_result}")
                    except OSError as e:
                        print("Error: %s - %s." % (e.filename, e.strerror))
                # Read the content of the config file
                with open(default_config_file_path, "r") as file:
                    config_content = file.read()

                # Find and replace the value of src_fn
                config_content = config_content.replace('src_fn           : 0', f'src_fn           : {this_file}')#it was difficult to write a pattern mattching codes for this lie so just ensure in the template file, src_fn is set to be 0 
                # Write the updated content back to the config file
                with open(new_config_file_path, "w") as file:
                    file.write(config_content)
                
                print(f"start to do posthoc analysis on {this_file}")
                subprocess.run(["fictrac.exe",target_config_file])
```
